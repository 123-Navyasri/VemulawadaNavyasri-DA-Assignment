```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file (replace 'path_to_your_file.csv' with your actual file path)
data = pd.read_csv('/content/insurance.csv')

# Scatter plot between age and expenses
plt.figure(figsize=(8, 6))
plt.scatter(data['age'], data['expenses'],
            c='blue', marker='+')
plt.title('Scatter Plot Between Age and Expenses')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.show()

```

```python
import matplotlib.pyplot as plt

#  Age and Smoking
plt.figure(figsize=(8, 6))
plt.scatter(data['age'], data['smoker'], c=data['smoker'].map({'yes': 'red', 'no': 'blue'}))
plt.title('Scatter Plot Between Age and Smoking')
plt.xlabel('Age')
plt.ylabel('smoker')
plt.show()
```

```python
# age between 20 and 40
filtered_data = data[(data['age'] >= 20) & (data['age'] <= 40)]

plt.figure(figsize=(8, 6))
plt.scatter(filtered_data['age'], filtered_data['smoker'], c=filtered_data['smoker'].map({'yes': 'red', 'no': 'blue'}))
plt.title('Scatter Plot Between Age(20-40) and Smoking')
plt.xlabel('Age')
plt.ylabel('smoker')
plt.show()
```

```python
smoker_data = filtered_data['smoker'].map({'yes': 1, 'no': 0})

plt.figure(figsize=(8, 6))

# Scatter plot where 'smoker' is on the x-axis (0 for 'no', 1 for 'yes')
plt.scatter(smoker_data,
            smoker_data,
            c=smoker_data.map({1: 'red', 0: 'blue'}),
            alpha=0.6, label='Smokers (Red), Non-Smokers (Blue)')

plt.title('Scatter Plot Between Smokers and Non-Smokers')
plt.xlabel('Smoking (0 = Non-Smoker, 1 = Smoker)')
plt.ylabel('Smokers vs Non-Smokers Representation')

# Adding grid for better visualization
plt.grid(alpha=0.3)
plt.show()
```

```python
# Count the number of smokers gender-wise
gender_smoking_count = pd.crosstab(data['sex'], data['smoker'])
gender_smoking_count.plot(kind='bar', stacked=True, figsize=(8, 6), color=['lightblue', 'pink'])
plt.title('Number of Smokers Gender-Wise')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter plot between Age and Expenses
plt.figure(figsize=(8, 6))
plt.scatter(data['age'], data['expenses'], c='green')
plt.title('Scatter Plot Between Age and Expenses')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter for teenage smokers
teen_smokers = filtered_data[(filtered_data['age'] >= 13) & (filtered_data['age'] <= 19) & (filtered_data['smoker'] == 'yes')]
num_teen_smokers = len(teen_smokers)
print(f'Number of teenage smokers: {num_teen_smokers}')

# Create scatter plot for smokers and non-smokers
plt.figure(figsize=(8, 6))

# Scatter plot for all individuals
plt.scatter(filtered_data['smoker'].map({'yes': 1, 'no': 0}),
            filtered_data['smoker'].map({'yes': 1, 'no': 0}),
            c=filtered_data['smoker'].map({'yes': 'red', 'no': 'blue'}),
            alpha=0.6, label='All Individuals')

# Highlight teenage smokers
plt.scatter(teen_smokers['smoker'].map({'yes': 1}),
            teen_smokers['smoker'].map({'yes': 1}),
            c='green', marker='+', label='Teen Smokers', s=100)

# Customize the plot
plt.title('Smokers and Non-Smokers (Teen Smokers Highlighted)')
plt.xlabel('Smoking (0 = Non-Smoker, 1 = Smoker)')
plt.ylabel('Smoking Status')
plt.legend()
plt.grid(alpha=0.3)

# Show the plot
plt.show()

```
